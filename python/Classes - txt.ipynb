```python
data = open('classes.txt')
```

```python
lista = []
lista2 = []
n = 0
```

```python
for linhas in data:
    lista.append(linhas)
```

```python
for linha in lista:
    n = n+1
    x = int(n/80)
    if linha.endswith('\n'):
        linha = linha.replace('\n',f':{x}\n')
    lista2.append(linha)   
        
```

```python
lista2
```

```python
arq = open('classes.txt','w')
arq.close()
```

```python
arq = open('classes.txt','r')
conteudo = arq.readlines()
conteudo.append(lista2)

arq = open('classes.txt','w')
arq.writelines(lista2)

arq.close()
```
